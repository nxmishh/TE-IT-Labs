```python
transactions = []
with open('Order1.csv','r') as file:
    for line in file:
        curr_trans = line.strip()
        tran_it = list(curr_trans.split(","))
        transactions.append(tran_it)
file.close()
```

```python
from mlxtend.preprocessing import TransactionEncoder
TE = TransactionEncoder()
x  =  TE.fit_transform(transactions)
x
```

```python
dataset = pd.DataFrame(x,columns = TE.columns_)
dataset
```
